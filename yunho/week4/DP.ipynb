```python
t = int(input())

for i in range(t):
    n, m = map(int, input().split())
    
    gold = list(map(int, input().split()))
    
    # 2차원 dp 테이블 초기화
    dp = []
    index = 0
    for i in range(n):
        dp.append(gold[index:index+m])
        index += m
        
    # dp[i][j] = array[i][j] + max(dp[i-1][j-1], dp[i][j-1], dp[i+1][j-1])
    
    # 다이나믹 프로그래밍 진행
    for j in range(1, m):
        for i in range(n):
            # 왼쪽 위에서 오는 경우
            if i == 0:
                left_up = 0
                
            # 왼쪽 아래에서 오는 경우
            else:
                left_up = dp[i-1][j-1]
            
            # 왼쪽에서 오는 경우
            if i == n-1:
                left_down = 0
            else:
                left_down = dp[i+1][j-1]
            
            left = dp[i][j-1]
            dp[i][j] = dp[i][j] + max(left_up, left_down, left)
            
    result = 0
    for i in range(n):
        result = max(result, dp[i][m-1])

    print(result)
```

```python
n = int(input())
dp = []

for _ in range(n):
    dp.append(list(map(int, input().split())))
    
# dp[i][j] = array[i][j] + max(dp[i-1][j], dp[i-1][j-1])

for i in range(1, n):
    for j in range(i+1):
        # 왼쪽 위에서 오는 경우
        if j == 0:
            up_left = 0
        else:
            up_left = dp[i-1][j-1]
        
        # 바로 위에서 오는 경우
        if j == i:
            up = 0
        else:
            up = dp[i-1][j]
        
        dp[i][j] = dp[i][j] + max(up_left, up)
        
print(max(dp[n-1]))
```
